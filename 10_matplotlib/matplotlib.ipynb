```python
# setup
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(-5,5,100)
f = np.sin(x)
```


<div style="max-width:66ch;">

# MATLAB style interface - don't use this approach

</div>

```python
plt.figure(dpi = 100)
plt.plot(x,f)
plt.title("sin(x)")
plt.xlabel("x")
plt.ylabel("y")
# plt.show() # needed for .py-script
```


<div style="max-width:66ch;">

## OOP approach

- fig is the canvas where you can put draw one or several axes
- an axes can contain axis, xlabel, ylabel, title, plot, scatter, histogram, bar, ... 

</div>

```python
# instantiates the objects fig and ax from Figure and AxesSubplot 
fig, ax = plt.figure(dpi=100),plt.axes()
ax.plot(x, f) # the ax object has a plot
# calling the method set from ax object to set the axis, labels and title
ax.set(xlim = (-6,6), ylim = (-1.5,1.5), xlabel="x", ylabel="y", title="$\sin{x}$");
# plt.show() # needed for .py-script
```


<div style="max-width:66ch;">

## Suplots 

</div>

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 4), dpi=200)  
ax[0].plot(x, np.sin(x))
ax[0].plot(x, np.cos(x))
ax[0].set(title="Plots", xlabel="x", ylabel="y")
ax[0].legend(("sin(x)", "cos(x)"))


ax[1].plot(x, np.cos(x))
ax[1].set(title="cos x", xlabel="x", ylabel="y")

# fig.show() needed for python script
```

some more charts

```python
fig, ax = plt.subplots(1, 2, dpi=150, figsize=(10, 4))

rng = np.random.RandomState(42)
# randn - draws random number from normal distribution
data_normal_dist = rng.randn(1000)

ax[0].hist(data_normal_dist, bins=30, alpha=0.5)
ax[0].set(title="Histogram", xlabel="x", ylabel="y")

size = 1000 * rng.rand(100)
sc = ax[1].scatter(
    x=rng.randn(100),
    y=rng.randn(100),
    c=rng.randn(100),
    alpha=0.3,
    cmap="viridis",
    s=size,
)
fig.colorbar(sc)
ax[1].set(title="Bubble chart", xlabel="x", ylabel="y")

fig.savefig("figures/beautiful_plots.png", transparent=True, facecolor="white")
# fig.show() # needed for .py-script
```
